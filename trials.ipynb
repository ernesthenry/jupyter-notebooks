```python
import numpy as np
import matplotlib.image as mpimg  # for reading in images

import matplotlib.pyplot as plt
import cv2  # computer vision library

%matplotlib inline
%matplotlib qt
# Read in image
image = cv2.imread('images/ernest.jpg')

# Print out the type of image data and its dimensions
print("This image is ", type(image),
     "with dimensions:", image.shape)
plt.imshow(image)
```

```python
%matplotlib inline

#Make the copy of the original image
image_copy = np.copy(image)

# Change from color to RGB (from BGR)
image_copy = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

#Display the image    
plt.imshow(image_copy)
```

```python
lower_blue = np.array([0,0,200]) 
upper_blue = np.array([250,250,255])
# Define the masked area
mask = cv2.inRange(image_copy, lower_blue, upper_blue)

# Vizualize the mask
plt.imshow(mask, cmap='gray')
```

```python
# Mask the image to let the person_image show through
masked_image = np.copy(image_copy)

masked_image[mask != 0] = [0, 0, 0]

# Display it!
plt.imshow(masked_image)
```

```python
# Load in a background image, and convert it to RGB 
background_image = cv2.imread('images/enable.jpg')
background_image = cv2.cvtColor(background_image, cv2.COLOR_BGR2RGB)

# Crop it to the right size (514x816)
crop_background = background_image[0:514, 0:816]

# Mask the cropped background so that the picture area is blocked
crop_background[mask == 0] = [0, 0, 0]

# Display the background
plt.imshow(crop_background)
```
